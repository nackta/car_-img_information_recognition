```python
from google.colab import drive
import os
import re
import pickle

drive.mount('/content/drive')

## img path 추출
root = "/content/drive/MyDrive/socar_hackathon/kkh/shooting_angle_to_car/car_data"
jpg_files = []
def search(dirname):
    filenames = os.listdir(dirname)
    for filename in filenames:
        full_filename = os.path.join(dirname, filename)
        ext = os.path.splitext(full_filename)[-1]
        if ext in ['.JPG', '.jpg']: 
            jpg_files.append(full_filename)
        else:
            search(os.path.join(dirname, filename))
search(root)

# shooting angle
v = []
h = []
for file_name in jpg_files:
    h.append(int(re.findall('_H[0-9]{1,3}', file_name)[0][2:]))
    v.append(int(re.findall('_V[0-9]{1,3}', file_name)[0][2:]))
angle = list(zip(h, v))

# shooting angle(minmaxscaling)
v = []
h = []
for file_name in jpg_files:
    h.append((int(re.findall('_H[0-9]{1,3}', file_name)[0][2:])-0)/345)
    v.append((int(re.findall('_V[0-9]{1,3}', file_name)[0][2:])-0)/60)
minmax_angle = list(zip(h, v))

print(len(jpg_files))
print(len(angle))
print(len(minmax_angle))

open_file = open("/content/drive/MyDrive/socar_hackathon/kkh/shooting_angle_to_car/jpg_files.pkl", "wb")
pickle.dump(jpg_files, open_file)
open_file.close()

open_file = open("/content/drive/MyDrive/socar_hackathon/kkh/shooting_angle_to_car/angle.pkl", "wb")
pickle.dump(angle, open_file)
open_file.close()

open_file = open("/content/drive/MyDrive/socar_hackathon/kkh/shooting_angle_to_car/minmax_angle.pkl", "wb")
pickle.dump(minmax_angle, open_file)
open_file.close()
```
